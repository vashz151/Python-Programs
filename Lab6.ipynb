```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('x axis label')

plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

plt.show()
```

```python
from scipy import stats, constants, linalg
from pylab import *
```

```python
dx=.001
X=np.arange(-2,2,dx)
Y= exp(-X**2)
Y /=(dx*Y).sum()

CY=np.cumsum(Y*dx)

plot(X,Y,'b--')
plot(X,CY,'r--')

plt.show()
```
